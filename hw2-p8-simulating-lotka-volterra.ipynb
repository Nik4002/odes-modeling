```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
def lotka_volterra(x, y, params):
    k = params["k"]
    l = params["l"]
    a = params["a"]
    b = params["b"]
    return (k*x - a*x*y, -l*y + b*x*y)
```

```python
def euler(f, params, x0, y0, Dt, tmax):
    x = [x0]
    y = [y0]
    t = [0]

    while t[-1] < tmax:
        t.append(t[-1] + Dt)        
        x.append(x[-1] + Dt * f(x[-1], y[-1], params)[0])
        y.append(y[-1] + Dt * f(x[-1], y[-1], params)[1])
    
    return (t, x, y)

initial_conditions = ((1, 1), (2, 3), (3, 5), (7, 10))
fig, ax = plt.subplots(1, 1 + len(initial_conditions), figsize = (20, 4))
for (n, (x0, y0)) in enumerate(initial_conditions):
    t, x, y = euler(lotka_volterra, {"k": 2, "l": 3, "a": 0.3, "b": 0.4}, x0, y0, 0.01, 10)
    ax[n].plot(t, x)
    ax[n].plot(t, y)
    ax[len(initial_conditions)].plot(x, y)

for a in ax[1:]:
    a.sharey(ax[0])
```

```python
def runge_kutta(f, params, x0, y0, Dt, tmax):
    x = [x0]
    y = [y0]
    t = [0]

    while t[-1] < tmax:
        t.append(t[-1] + Dt)
        k1_x, k1_y = f(x[-1], y[-1], params)
        k2_x, k2_y = f(x[-1] + Dt*k1_x/2, y[-1] + Dt*k1_y/2, params)
        k3_x, k3_y, = f(x[-1] + Dt*k2_x/2, y[-1] + Dt*k2_y/2, params)
        k4_x, k4_y = f(x[-1] + Dt*k3_x, y[-1] + Dt*(k3_y), params)
        x.append(x[-1] + (Dt/6) * (k1_x + 2*k2_x + 2*k3_x + k4_x))
        y.append(y[-1] + (Dt/6) * (k1_y + 2*k2_y + 2*k3_y + k4_y))
    
    return (t, x, y)

initial_conditions = ((1, 1), (2, 3), (3, 5), (7, 10))
fig, ax = plt.subplots(1, 1 + len(initial_conditions), figsize = (20, 4))
for (n, (x0, y0)) in enumerate(initial_conditions):
    t, x, y = runge_kutta(lotka_volterra, {"k": 2, "l": 3, "a": 0.3, "b": 0.4}, x0, y0, 0.01, 10)
    ax[n].plot(t, x)
    ax[n].plot(t, y)
    ax[len(initial_conditions)].plot(x, y)

for a in ax[1:]:
    a.sharey(ax[0])
```
