```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
def lorenz(x, y, z, params):
    s = params["s"]
    r = params["r"]
    b = params["b"]

    return (s * (y - x), x * (r - z) - y, x * y - b * z)
```

```python
def euler(f, params, y1_0, y2_0, y3_0, Dt, tmax):
    y1 = [y1_0]
    y2 = [y2_0]
    y3 = [y3_0]
    t = [0]

    while t[-1] < tmax:
        t.append(t[-1] + Dt)
        Dv = f(y1[-1], y2[-1], y3[-1], params)
        y1.append(y1[-1] + Dt * Dv[0])
        y2.append(y2[-1] + Dt * Dv[1])
        y3.append(y3[-1] + Dt * Dv[2])
    
    return (t, y1, y2, y3)

Dt = 0.01
tmax = 20
params = {"s": 10, "r": 28, "b": 8/3}

```

```python
initial_conditions = ((-1, -1, -1), (1, 1, 1))
x_plot = plt.figure().add_subplot()
ax = plt.figure().add_subplot(projection='3d')

for (n, (y1_0, y2_0, y3_0)) in enumerate(initial_conditions):
    t, y1, y2, y3 = euler(lorenz, params, y1_0, y2_0, y3_0, Dt, tmax)
    x_plot.plot(t, y1)
    ax.plot(y1, y2, y3, linewidth = 0.5)
```

```python
initial_conditions = ((1, 1, 1), (1.001, 1.001, 1.001))
x_plot = plt.figure().add_subplot()
phase_plot = plt.figure().add_subplot(projection='3d')

for (n, (y1_0, y2_0, y3_0)) in enumerate(initial_conditions):
    t, y1, y2, y3 = euler(lorenz, params, y1_0, y2_0, y3_0, Dt, tmax)
    x_plot.plot(t, y1)
    phase_plot.plot(y1, y2, y3, linewidth = 0.5)
```
